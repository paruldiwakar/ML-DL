```python
from sklearn.cluster import KMeans
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

```python

img = cv2.imread("/Users/paruldiwakar/Desktop/EXO/xexo_kai.jpg")
im= cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.imshow(im)
plt.show()
```

```python
all_pixels=im.reshape((-1,3))
```

```python
k=4
km=KMeans(n_clusters=k)
km.fit(all_pixels)
```

```python
centers=km.cluster_centers_
```

```python
centers= np.array(centers, dtype="uint8")
```

```python
print(centers)
```

```python
i=1
plt.figure( figsize=(4,2))

colors=[]

for each_col in centers:
    plt.subplot(1,5,i)
    i+=1
    
    colors.append(each_col)
    
    a=np.zeros((100,100,3),dtype=int)
    
    a[:,:,:]=each_col
    
    plt.axis('off')
    plt.imshow(a)
    

plt.show()    
    
```

```python
new_img = np.zeros(im.shape, dtype="uint8").reshape((-1, 3))
```

```python
new_img.shape
```

```python
km.labels_
```

```python
# assign most dominating color according to group it belongs
for ix in range(new_img.shape[0]):
    new_img[ix] = centers[km.labels_[ix]]
```

```python
new_img  =  new_img.reshape(im.shape)
```

```python
plt.imshow(im)
plt.show()
plt.imshow(new_img)
plt.show()
```
